```python
import pandas as pd

def check_user_in_excel(surname, department):
    df = pd.read_csv("jt-ventures.csv")
    
    user = df[(df['SURNAME'] == surname) & (df['DEPARTMENT'] == department)]
    
    if not user.empty:
            print("User found in the excel file")
    else:
        print("User not found in the excel file ")
        
surname = input("Please input your surname: ")

department = input("What is your department: ")

check_user_in_excel(surname, department)

```
